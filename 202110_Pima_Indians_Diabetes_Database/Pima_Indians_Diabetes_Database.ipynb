About : 
- The datasets consist of several medical predictor (independent) variables and one target (dependent) variable, Outcome. Independent variables include the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

Column Explain:
- Pregnancies : Number of times pregnant
- Glucose : Plasma glucose concentration a 2 hours in an oral glucose tolerance test
(경구 포도당 내성 검사에서 2시간 동안의 혈장 포도당 농도)
- BloodPressure : Diastolic blood pressure (mm Hg)
- SkinThickness : Triceps skin fold thickness (mm)
(삼두근 피부 주름 두께(mm))
- Insulin : 2-Hour serum insulin (mu U/ml)
- BMI : Body mass index (weight in kg/(height in m)^2)
- DiabetesPedigreeFu : Diabetes pedigree function
(당뇨병 가계도 기능)
- Age
- Outcome : Class variable (0 or 1) 268 of 768 are 1, the others are 0

```python
# df['Age'].value_counts().sort_index(ascending = True)
plt.figure(figsize=(15,5))
sns.countplot(df['Age'])
plt.xticks(rotation = 30)
plt.title("Age Count", fontsize = 16)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(df['Pregnancies'])
plt.xticks(rotation = 30)
plt.title("Pregnency Times", fontsize = 16)
plt.show()
```

```python
plt.figure(figsize =(15,5))
sns.barplot(data=df, x='Pregnancies', y='Age')
plt.xlabel("Pregnency Times")
plt.xticks(rotation = 30)
plt.title("Relation between Age and Pregnency Times", fontsize=16)
plt.show()
```

```python
plt.figure(figsize =(15,8))

plt.subplot(2,2,1)
sns.scatterplot(data=df, x='BMI', y='Age', hue='Outcome')

plt.subplot(2,2,2)
sns.scatterplot(data=df, x='BloodPressure', y='Age', hue='Outcome')

plt.subplot(2,2,3)
sns.scatterplot(data=df, x="Glucose", y='Age', hue = 'Outcome')

plt.subplot(2,2,4)
sns.scatterplot(data=df, x='Insulin', y='Age', hue ='Outcome')

plt.tight_layout()
plt.show()
```

```python
df_diabetic = df.copy()
df_diabetic['Outcome'].replace(0,'Healthy', inplace=True)
df_diabetic['Outcome'].replace(1,'Diabetic', inplace=True)

df_diabetic.head()
```

```python
df_diabetic['Outcome'].value_counts().sort_index()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(df_diabetic['Outcome'])
plt.xlabel('Diabetes or Healthy')
plt.title("Diabeties Count", fontsize=16)
plt.show()
```

```python
out_df= pd.DataFrame(df.groupby('Outcome')['Outcome'].count())
colors = ["green","red"]
plt.pie(out_df['Outcome'], labels=['healthy', 'Diabetic'], colors=colors, autopct ="%.0f%%", 
        radius=2, explode=(0,0.1))
plt.legend()
plt.show()
```

```python
sns.pairplot(df, hue='Outcome', corner= True)
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.suptitle('Diabetes Age and Pregenency')
plt.subplot(1,2,1)
sns.violinplot(data=df, y='Age', x='Outcome')

plt.subplot(1,2,2)
sns.violinplot(data=df, y='Pregnancies', x='Outcome')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap", fontsize = 16)
plt.xticks(rotation=45)
plt.show()
```

### 2. Modeling

```python
plt.figure(figsize = (12,8))
plt.suptitle("Plot Outliars", fontsize=16)

index = 1
for col in df.columns.tolist():
    try:
        plt.subplot(4,2,index)
        if index < 8:
            index += 1
        
        sns.boxplot(data=df, x=col)
    except ValueError as e:
        print(type(e), e)

plt.tight_layout()
plt.show()

```
